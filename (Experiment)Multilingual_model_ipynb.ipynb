```python
def translate_df(df, n=200, lang='en'):
  """  
  If you receive adata frame (df) as input and specify the column (col) to be translated by google translator as a string type,
  The Korean-English translated string value is appended to the dataframe.
  df = input dataframe
  col = column name to translate (string type)
  lang = default is set to English (en) and the language setting can be changed 
  """
  import random
  from tqdm import tqdm
  
  meta_NoneEng = meta_data3[meta_data3['LANGUAGE'].apply(lambda x: x !=['eng'])]

  test_indices = random.sample(list(meta_NoneEng.index), n)
  meta_test = meta_NoneEng.loc[test_indices]

  test_title = meta_test['TITLE'].values
  test_abstract = meta_test['ABSTRACT'].values

  test_title_trans = [translator.translate(i, src='auto',dest='en').text for i in tqdm(test_title)]
  # test_abstract_trans = [translator.translate(i[0], src='auto',dest='en').text for i in tqdm(test_abstract)]

  test_abstract_trans = []
  for i in tqdm(test_abstract):
    try:
      test_abstract_trans.append(translator.translate(i[0], src='auto',dest='en').text)
    except:
      test_abstract_trans.append('')


  meta_test['TITLE'] = test_title_trans
  meta_test['ABSTRACT'] = test_abstract_trans
  return meta_test                                                                      # The data frame converted into English is appended to the existing data frame and returned
```
