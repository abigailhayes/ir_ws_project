## **Process of Translating from auto detection to English(not completed)**

```python
def translate_df(df, col, lang='en'):
  """  
  If you receive adata frame (df) as input and specify the column (col) to be translated by google translator as a string type,
  The Korean-English translated string value is appended to the dataframe.
  df = input dataframe
  col = column name to translate (string type)
  lang = default is set to English (en) and the language setting can be changed 
  """
 
  translator = googletrans.Translator()
  df['eng'] = [translator.translate(i, src='auto',dest=lang).text for i in df[col]]   # Auto-detect the original data frame column value and translate it into English
   
  tmp1 = df.drop(['eng'], axis=1)                                                    # Create two temporary dataframes and merge the original and English translation dataframes
  tmp2 = df.drop([col], axis=1)
  tmp2.rename(columns={'eng':'data'}, inplace=True)
  result = pd.concat([tmp1,tmp2], ignore_index=True)
  result = result.drop_duplicates()
  return result                                                                      # The data frame converted into English is appended to the existing data frame and returned
```
